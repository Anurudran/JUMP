```python
import names
import uuid
import random
import datetime
import pandas as pd
```

```python
def list_ids(num):
  id_list = []
  for i in range(num):
    id_list.append(uuid.uuid4())
  return id_list

def list_names(num):
  name_list = []
  for i in range(num):
    name_list.append(names.get_full_name())
  return name_list

def list_salaries(num):
  salary_list = []
  for i in range(num):
    salary_list.append(random.randrange(50000, 100000))
  return salary_list

def list_departments(num):
  department_list = []
  departments = [
      "Sales",
      "HR",
      "IT",
      "R & D"
  ]
  for i in range(num):
    department_list.append(random.choice(departments))
  return department_list

def list_starts(num):
  start_list = []
  for i in range(num):
    start_list.append(datetime.datetime.now())
  return start_list

def list_employed(num):
  employed_list = []
  for i in range(num):
    employed_list.append(random.choice([True, False]))
  return employed_list
```

```python
id_series = pd.Series(list_ids(1000))
names_series = pd.Series(list_names(1000))
salaries_series = pd.Series(list_salaries(1000))
department_series = pd.Series(list_departments(1000))
start_series = pd.Series(list_starts(1000))
employed_series = pd.Series(list_employed(1000))
```

```python
print(id_series.head())
```

```python
names_series.index = id_series
salaries_series.index = id_series
department_series.index = id_series
start_series.index = id_series
employed_series.index = id_series
```

```python
print(department_series.head(20))
```

```python
data = {'Employee ID': id_series,
        'Employee Name': names_series,
        'Employee Salary': salaries_series,
        'Employee Department': department_series,
        'Employee Start Date': start_series,
        'Currently Employed': employed_series}
df = pd.DataFrame(data)
df
```

```python
min_salary = df["Employee Salary"].min()
max_salary = df["Employee Salary"].max()
mean_salary = df["Employee Salary"].mean()
sum_salary = df["Employee Salary"].sum()
print(f"min is {min_salary}, max is {max_salary}, Average is {mean_salary}, sum is {sum_salary}")
```

```python
print(df.groupby("Employee Department")["Employee Salary"].min())
print(df.groupby("Employee Department")["Employee Salary"].max())
print(df.groupby("Employee Department")["Employee Salary"].sum())
print(df.groupby("Employee Department")["Employee Salary"].mean())

```
